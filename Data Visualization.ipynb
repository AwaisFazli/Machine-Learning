```python
x1 = [20,40,80,100]
y1 = [10,20,30,40]

x2 = [30 ,40, 50, 60]
y2 = [30,50,10,40]

style.use("ggplot")

plt.plot(x1, y1, "c", label = "Line1", linewidth = 2)
plt.plot(x2, y2, "r", label = "line2", linewidth = 2)

plt.legend()

plt.grid(False, color = "g")

plt.show()
```
