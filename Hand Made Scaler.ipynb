```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler  , RobustScaler, Normalizer, FunctionTransformer
import math
```

```python
df1 = pd.DataFrame({
                   "x1" : np.random.normal(0, 5 ,1000),
                   "x2" : np.random.normal(5, 5 ,1000),
                   "x3" : np.random.normal(2, 5 ,1000),
})
```

```python
def mapper(x):
  listed = pd.DataFrame()
  for keys , data in x.iteritems():
    for i in data:
      a = data.max()
      b = data.min()
    z = (data - b)/(a - b)
    col = keys
    listed[col] = z
  return listed
```

```python
ctd = FunctionTransformer(mapper , validate = False)
dft = ctd.fit_transform(df1)
```
