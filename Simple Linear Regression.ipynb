```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/Collab DataSets/headbrain.csv")
dataset.head()
```

```python
size = dataset["Head Size(cm^3)"]
weight = dataset["Brain Weight(grams)"]
```

```python
plt.scatter(size , weight)
```

```python
x = np.array(size).reshape([-1,1])
y = np.array(weight).reshape([-1,1])
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
mse = mean_squared_error(x,y)
print(f"The Value Of MSE is {math.sqrt(mse)} and Model Score Is {model.score(x,y)}")
```

```python
plt.scatter(x, y, color = "orange")
plt.plot(x, model.predict(x), color = "k")
```
