```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import math
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mhuzaifadev/mlzero_to_hero/main/06_Polynomial_Regression/Drugs_Performance.csv")
df.head()
```

```python
X = df[["Age"]].values
Y = df[["Drugs_Perf"]].values
```

```python
plt.scatter(X,Y)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y, test_size = 0.2)
```

```python
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X_train)
X_poly
```

```python
column=["x1", "x2", "x3"]
db = pd.DataFrame(X_poly ,columns=column)
plt.scatter(db.x1 , Y_train)
plt.scatter(db.x2 , Y_train)
plt.scatter(db.x3 , Y_train)
```

```python
model = LinearRegression()
model.fit(X_poly, Y_train)
```

```python
plt.scatter(X,Y, color = "orange")
plt.plot(X, model.predict(poly.fit_transform(X)), color = "k")
```

```python
X_test_poly = poly.fit_transform(X_test)
X_predicted = model.predict(X_test_poly)
```

```python
mse = mean_squared_error(Y_train, model.predict(poly.fit_transform(X_train)))
```

```python
print(f"The Mean Squared Error Is {math.sqrt(mse)}")
```
