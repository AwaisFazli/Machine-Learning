```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
digits = datasets.load_digits()

images_and_labels = list(zip(digits.images, digits.target))
```

```python
 n_samples = len(digits.images)

 data = digits.images.reshape(n_samples, -1)
 data
```

```python
traintestsplit = int(n_samples * 0.75)

feature_train_data = data[: traintestsplit]
target_train_data  = digits.target[: traintestsplit]
```

```python
model = svm.SVC()
model.fit(feature_train_data, target_train_data)
```

```python
target_test = digits.target[traintestsplit :]

predicted = model.predict(data[traintestsplit :])
```

```python
print(confusion_matrix(target_test, predicted))
print(accuracy_score(target_test, predicted))
```

```python
sns.heatmap(confusion_matrix(target_test, predicted), annot=True, cmap="Blues")
```
