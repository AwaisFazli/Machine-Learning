```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/drive/MyDrive/Collab DataSets/50_Startups.csv")
df.head()
```

```python
plt.scatter(df["R&D Spend"], df["Profit"])
plt.scatter(df["Administration"], df["Profit"])
plt.scatter(df["Marketing Spend"], df["Profit"])
```

```python
x = df[["R&D Spend","Administration","Marketing Spend","State"]].values
y = df[["Profit"]].values
```

```python
st = ColumnTransformer([("State",OneHotEncoder(), [3])], remainder="passthrough")
x = st.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
for y in range(len(y_pred)):
  print(f"The Actual Value Is {y_test[y]} and Predicted Value Is {y_pred[y]}")
```

```python
randmodel = LinearRegression()
randmodel.fit(randd, randprofit)

adminmodel = LinearRegression()
adminmodel.fit(admin, adminprofit)

Marketingmodel = LinearRegression()
Marketingmodel.fit(Marketing, Marketingprofit)
```

```python
randpredict = randmodel.predict(randdtest)
adminpredict = adminmodel.predict(admintest)
Marketingpredict = Marketingmodel.predict(Marketingtest)
```

```python
plt.plot(randdtest, randpredict)
plt.plot(adminpredict, admintest)
plt.plot(Marketingpredict, Marketingtest)

plt.scatter(randprofittest, randdtest)
plt.scatter(adminprofittest, admintest)
plt.scatter(Marketingprofittest, Marketingtest)
```
